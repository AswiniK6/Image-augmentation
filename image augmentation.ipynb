```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from PIL import Image
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='wrap')
```

```python
img=load_img(r"C:\Users\Hp\Downloads\AI-images-show-Elon-Musk-as-an-Indian-groom-wearing-a-sherwani.jpg")
```

```python
img
```

```python
x = img_to_array(img)  
x = x.reshape((1,) + x.shape)  


i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir=r"E:\data imagation", save_prefix='elon musk', save_format='jpeg'):
    i += 1
    if i > 40:
        break 
```
